### Background
The primitive equations:
1. Horizontal acceleration equation (F=ma):
$$ \frac{D}{Dt} \vec V_h = -f \hat k \times \vec V_h - \vec \nabla_p \Phi +\vec F_r  $$
or, in component form, 
$$ \frac{Du}{Dt} = fv  - \frac{\partial \Phi}{\partial x} +F $$
$$ \frac{Dv}{Dt} = -fu - \frac{\partial \Phi}{\partial y} +G $$

1. Vertical acceleration equation (aka Hydrostatic balance): 
$$ \frac {\partial \Phi}{\partial p} = - \alpha \equiv -\frac{1}{\rho} = \frac{RT}{p} $$
1. Mass continuity: 
$$ 0 = \frac {\partial u}{\partial x} + \frac {\partial v}{\partial y}  + \frac {\partial \omega}{\partial p} $$
1. First Law of thermodynamics: 
$$ \frac {D_h T}{Dt} + \omega \frac {\partial \theta}{\partial p} = J/C_p $$


Now suppose the flow is in geostrophic balance (dV/dt=0, F=G=0). This defines the geostrophic wind:  
$$ fv_g = +\frac{\partial \Phi}{\partial x}$$
$$ fu_g = -\frac{\partial \Phi}{\partial y}$$


Taking the p derivative, and noticing that $\frac{\partial \Phi}{\partial p} = g\frac{\partial Z}{\partial p}$ is the *geometric thickness* (dZ) of a layer with mass (dp/g), 
$$ \frac {\partial}{\partial p}v_g = +g/f \frac{\partial \frac {\partial Z}{\partial p}}{\partial x}$$
$$ \frac {\partial}{\partial p}u_g = -g/f \frac{\partial \frac {\partial Z}{\partial p}}{\partial y}$$

The horizontal pattern of thickness is the *pattern of temperature*, as you know. But actually plugging in the hydrostatic equation above for $\frac{\partial \Phi}{\partial p}$ is complicated, because p is in the denominator of the right hand side. ## Exercise 4 answer: 

Thermal wind can be seen to flow parallel to the temperature contours above, whereas the temperatue gradient would have to be perpendicular to the temperature contours. Since the temperature gradient and the thermal wind are therefore perpendicular to one another, their dot product would be zero when both values were placed in the advection equation. This shows that the thermal wind does not advect temperature. 
When comparing the two heights temperature advection equations above, the dot product between the temp gradient and thermal wind would again be zero. This would leave only the dot product of the geostrophic wind at each height and the temperature gradient. These dot products would result in the same value for each height, i.e. temperature advections at each height by geostrophic winds are equal.
