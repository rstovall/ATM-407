### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= \frac{\partial}{\partial x} (U + c_1x + c_2y) + \frac{\partial}{\partial y} (V + c_3x + c_4y) = \frac{\partial U}{\partial x} + c_1\frac{\partial x}{\partial x} + c_2\frac{\partial y}{\partial x} + \frac{\partial V}{\partial y} + c_3\frac{\partial x}{\partial y} + c_4\frac{\partial y}{\partial y} = c_1 + c_4 $
* $ \zeta_{rel} = \frac{\partial V}{\partial x} - \frac{\partial U}{\partial y} = \frac{\partial}{\partial x}(V + c_3x + c_4y) - \frac{\partial}{\partial y}(U + c_1x + c_2y) = \frac{\partial V}{\partial x} + c_1\frac{\partial x}{\partial x} + c_4\frac{\partial y}{\partial x} - \frac{\partial U}{\partial y} - c_1\frac{\partial x}{\partial y} - c_2\frac{\partial y}{\partial y} = c_3 - c_2 $
* $ def1 = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} = \frac{\partial}{\partial x}(U + c_1x + c_2y) - \frac{\partial}{\partial y}(V + c_3x + c_4y) = \frac{\partial U}{\partial x} + c_1\frac{\partial x}{\partial x} + c_2\frac{\partial y}{\partial x} + \frac{\partial V}{\partial y} + c_3\frac{\partial x}{\partial y} + c_4\frac{\partial y}{\partial y} = \frac{\partial U}{\partial x} + \frac{\partial V}{\partial y} + c_1 + c_4 = (div) + c_1 + c_4 = c_1 + c_4 $
* $ def2 = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y} = \frac{\partial}{\partial x}(V + c_3x + c_4y) - \frac{\partial}{\partial y}(U + c_1x + c_2y) = \frac{\partial V}{\partial x} + c_3\frac{\partial x}{\partial x} + c_4\frac{\partial y}{\partial x} - \frac{\partial U}{\partial y} - c_1\frac{\partial x}{\partial y} - c_2\frac{\partial y}{\partial y} = \frac{\partial V}{\partial x} - \frac{\partial U}{\partial y} + c_3 - c_2 = (\zeta_{rel}) + c_3 - c_2 = c_3 - c_2 $

## Exercise 2: Create and display the basic fields
* First, execute a code cell setting up a coordinate system (x,y) from -10 to 10
* Next, execute the cell defining the basic "pure" div, vort, def1, def2, and U,V fields 
* Finally, display them. The first one has been done for you. You complete the work. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

## Answer to Exercise 4: 

Vorticity (positive here) plus straight flow can be described as a cyclonic flow, whose propogration path is determined by the strength of the vector field (U).    

At lower levels, cyclonic flow associated with positive vorticity will cause southerly winds to the east leading to warm air horizontal advection. Northerly winds will be created to the west, leading to cold air horizontal advection. These effects propogate the initial low level positive vorticity towards the warm air advection to the east. 

At upper levels, positive vorticity causes cyclonic circulation. To the west of this circulation, air will be advected southward. To the east, advection will occur in a northward direction. The combination of these effects propogrates the initial positive vorticity to the west.
