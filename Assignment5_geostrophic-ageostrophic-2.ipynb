### Background and math

You will decompose the total wind into geostrophic and ageostrophic aspects: $\vec V = \vec V_g + \vec V_a$. 

* Horizontal acceleration equations (F=ma) for inviscid flow:
$$ \frac{du}{dt} = - \frac{\partial \Phi}{\partial x} + fv $$

$$ \frac{dv}{dt} = - \frac{\partial \Phi}{\partial y} - fu $$

* The geostrophic wind is defined as *the wind that would occur if the LHS were zero*: 

$$ 0 = - \frac{\partial \Phi}{\partial x} + fv_g $$

$$ 0 = - \frac{\partial \Phi}{\partial y} - fu_g $$

## Exercise 1 Answer:

$$ \frac{\partial}{\partial y}(fv_g = \frac{\partial \Phi}{\partial x}) $$
$$ v_g\frac{\partial f}{\partial y} + f\frac{\partial v_g}{\partial y} = \frac{\partial \Phi}{\partial x \partial y}$$
$$ \frac{\partial v_g}{\partial y} = \frac{1}{f}\frac{\partial \Phi}{\partial x \partial y} - \frac{v_g}{f}\frac{\partial f}{\partial y} $$

$$ \frac{\partial}{\partial x}(fu_g = -\frac{\partial \Phi}{\partial y}) $$
$$ u_g\frac{\partial f}{\partial x} + f\frac{\partial u_g}{\partial x} = -\frac{\partial \Phi}{\partial x \partial y}$$
$$ \frac{\partial u_g}{\partial x} = -\frac{1}{f}\frac{\partial \Phi}{\partial x \partial y} - \frac{u_g}{f}\frac{\partial f}{\partial y} $$

$$ \frac{\partial u_g}{\partial x} + \frac{\partial v_g}{\partial y} = -\frac{1}{f}\frac{\partial \Phi}{\partial x \partial y} - \frac{u_g}{f}\frac{\partial f}{\partial x} + \frac{1}{f}\frac{\partial \Phi}{\partial x \partial y} - \frac{v_g}{f}\frac{\partial f}{\partial y} = - \frac{u_g}{f}\frac{\partial f}{\partial x} - \frac{v_g}{f}\frac{\partial f}{\partial y} = -\frac{v_g}{f}\frac{\partial f}{\partial y} $$

$$ \beta = df/dy = 0 $$
$$ -\frac{v_g}{f}\frac{\partial f}{\partial y} = 0 $$


#### Key points: the geostrophic wind is boring for 2 reasons: because it is not connected to vertical velocity $\omega$, and also because it would never change with time since its definition sets dV/dt to zero. The implication is that ageostrophic wind, despite its relatively small magnitude, is especially *interesting*. However, the geostrophic wind is adeeequate to approximate the direction and speed of advection of various scalar fields (temperature, vorticity) in the atmosphere.

## Answer to exercise 3: Acceleration-related ageostrophic winds 

In this ridge, ageostrophic wind and true wind flow in the same direction. North of the western US high pressure system (below image), parcels flowing over Alberta are accelerating as they pass through the ridge (where contours are getting closer together). The pressure gradient force on a vector there would be to the north/northeast, pointing away from southward high pressure. The Coriolis force on moving parcel in the Northern Hemisphere is always to the right, which for one of the vectors depicted in the ridge here would be to the south. 
A perfect balance between these two forces would lead to a geostrophic flow. Clearly, since they do not point perfectly in opposite directions, they are not balanced, and the resultant true wind is not geostrophic. A perfectly balanced geostrophic wind would not accelerate, but the ageostrophic wind is equal to the horizontal acceleartion since it is by definition the difference between the true wind and perfectly geostrophic wind.
